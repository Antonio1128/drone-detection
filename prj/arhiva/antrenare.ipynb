```python
from ultralytics import YOLO
from ultralytics.data.utils import download

# 1. Descarcă automat setul de date oficial coco8 în folderul prj
download(url="https://github.com/ultralytics/assets/releases/download/v8.1.0/coco8.zip", dir=".")

# 2. Încarcă modelul gol
model = YOLO("yolov8n.pt")

# 3. Pornește antrenarea
model.train(data="coco8/data.yaml", epochs=5, imgsz=640)
```
