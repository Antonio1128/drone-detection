```python
# 2. Verifică GPU
import torch
print('GPU disponibil:', torch.cuda.is_available())
print('GPU:', torch.cuda.get_device_name(0) if torch.cuda.is_available() else 'Niciun GPU')
```

```python
# 3. Uploadează dataset-ul
# Comprima folderul prj/datasets/drones ca drones.zip si uploadeaza-l in Colab
from google.colab import files
uploaded = files.upload()  # selecteaza drones.zip
```

```python
# 4. Dezarhivează
!unzip -q drones.zip -d datasets/drones
!ls datasets/drones
```

```python
# 5. Creaza data.yaml
yaml_content = """
path: /content/datasets/drones
train: images/train
val: images/val
test: images/test
nc: 1
names: ['drone']
"""
with open('data.yaml', 'w') as f:
    f.write(yaml_content)
print('data.yaml creat!')
```

```python
# 6. Antreneaza modelul
from ultralytics import YOLO

model = YOLO('yolov8s.pt')  # small - mai bun decat nano
results = model.train(
    data='data.yaml',
    epochs=100,
    imgsz=640,
    batch=16,
    device=0,  # GPU
    project='runs/detect',
    name='drone_v2',
)
print('Antrenare terminata!')
```

```python
# 7. Descarca modelul antrenat
from google.colab import files
files.download('runs/detect/drone_v2/weights/best.pt')
print('best.pt descarcat!')
```
